```python
# 문제 1) Series 만들기
# 다음 과일 이름으로 pd.Series를 만들어 fruits 변수에 저장하시오
# : ['사과', '바나나', '포도']
import pandas as pd
fruits = pd.Series(['사과','바나나','포도'])
fruits
```

```python
# 문제 2) DataFrame 만들기 (리스트로 생성)
# 다음 데이터를 이용해 아래와 같은 컬럼을 가진 df를 생성하시오.

# 컬럼: 상품, 가격, 재고
# 데이터:
# 상품: ['키보드', '마우스', '모니터']
# 가격: [30000, 15000, 200000]
# 재고: [10, 25, 7]
df = pd.DataFrame({
        '상품':['키보드','마우스','모니터'],
        '가격':[30000,15000,200000],
        '재고':[10,25,7]
})
df
```

```python
# 문제 3) 단일 컬럼 선택 (Series vs DataFrame)
# 문제 2에서 만든 df로부터 상품 컬럼을 (1) Series로,
# (2) DataFrame으로 각각 선택해 prod_s, prod_df에 저장하시오.
prod_s = df['상품']
prod_s
prod_df = df[['상품']]
prod_df
```

```python
# 문제 4) 여러 컬럼 선택
# 문제 2의 df에서 상품과 가격 컬럼만 선택하여
# price_view 변수에 저장하시오.
price_view = df[['상품','가격']]
price_view
```

```python
# 문제 5) loc로 조건 필터링
# 문제 2의 df에서 가격이 20000 이상인 행만 loc로 필터링하여
# expensive에 저장하시오.
# 단, 컬럼은 상품, 가격만 선택하시오.
expensive = df.loc[df['가격'] >=20000, ['상품','가격']]
expensive
# df['가격'] >= 20000 조건을 만족하는 행만 가져오는데,
# 이때 원래 DataFrame의 인덱스가 그대로 따라옵니다.
# 원래 DataFrame에서 "키보드"는 인덱스 0, "모니터"는 인덱스 2였기 때문에
# 그대로 0, 2로 표시된 거예요.
```

```python
# 문제 6) loc로 값 수정
# 문제 2의 df에서 상품이 '마우스'인 행의 가격을 17000으로 수정하시오.
df.loc[df['상품']=='마우스','가격'] = 17000
df[1:2]
```

```python
# 문제 7) iloc로 부분 선택
# 문제 2의 df에서 첫 2행과 앞의 2개 컬럼만 iloc으로 선택해
# sub_iloc에 저장하시오.
sub_iloc = df.iloc[0:2,0:2]
sub_iloc
```

```python
# 문제 8) 새 컬럼 추가
# 문제 2의 df에 **부가세 포함 가격(10% 가산)**을 정수형으로 계산하여
# 부가세포함가 컬럼으로 추가하시오.
df['부가세포함가'] = (df['가격'] * 1.10).astype(int)
df
```

```python
# 문제 9) 정렬
# 문제 2의 df를 가격 기준 내림차순으로 정렬한 결과를
# df_sorted에 저장하시오.
df_sorted = df.sort_values(by='가격',ascending=False)
df_sorted
```

```python
# 8-1) 정렬
# 가격 오름차순
df_sorted_asc = df.sort_values(by='가격', ascending=True)
# df_sorted_asc
# 가격 내림차순
df_sorted_desc = df.sort_values(by='가격', ascending=False)
display(df_sorted_asc,df_sorted_desc)
```
