```python
import pandas as pd

df = pd.DataFrame({
    "이름": ["철수", "영희", "민수", "지현", "수빈", "현우", "하늘","버섯"],
    "국어": [120, 90, 70, 95, 100, 50, 80,50],
    "영어": [78, 88, 82, -5, 85, 70, 95,90],
    "수학": [92, 85, 60, 98, 90, 75, 88,-10]
})
df
```

```python
# 1) 데이터 샘플 확인
#앞에서부터 n개
df.head(3) #전달된 인자값만큼의 행 출력
```

```python
# 뒤에서부터 n개
df.tail(2) #전달된 인자만큼의 행 뒤에서부터 출력
```

```python
# 임의 샘플
df.sample(3) #랜덤한 행 전달된 인자만큼 출력
```

```python
# 2) 데이터 구조 확인
# 데이터 크기
df.shape
# 출력값 (7,4) 7행 4컬럼
```

```python
# 컬럼 타입/결측치 정보
df.info()
#int 수치형 데이터 / object 범주형 데이터
# 수치형 데이터는 시각화 하기 용이
# 범주형 데이터는 까다로움
# 범주형 / 수치형을 시각화 처리하는 방식이 나뉘어짐
```

```python
# 3) 기초 통계 및 상관관계
# 기초 통계 요약
df.describe()
# df.describe(include='O')// <<<<< 범주형 데이터 확인
# count - 전체 데이터의 개수
# mean - 평균?
# std - 표준편차 (데이터의 분포 정도)
# min - 해당 컬럼의 최소값
# 25% - Q1 사분위수
# 50% - Q2
# 75% - Q3
# max - 최대값
```

```python
# 상관관계
df.corr(numeric_only=True)

# 1은 자기 자신을 의미
# 상관관계가 1에 가까울 수록 좋지 않을 가능성이 높다
# 1에 가까운 것들중에 하나는 빼던가 하는 방향을 고려할 수 있으면 좋음
```

```python
# 4) 범주형 데이터 탐색
# 예시 데이터 (학급 & 성별)
df_class = pd.DataFrame({
    "이름": ["철수","영희","민수","지현","수빈","현우","하늘"],
    "학급": ["A","A","B","B","A","C","C"],
    "성별": ["M","F","M","F","F","M","F"]
})
df_class
```

```python
# 항목 종류
df_class['학급'].unique()
# df_class['성별'].unique()
```

```python
# 항목별 개수
df_class['학급'].value_counts()
# df_class['성별'].value_counts()
```

```python
# 5) 범주형 데이터 요약
df_class.describe(include="O")
```
