```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


# 데이터 준비 (시간에 따른 주가 변화 가정)
# x = np.linspace(0, 10, 100) # 0부터 10까지 100개의 점
# y = np.sin(x) + np.random.normal(0, 0.1, 100)

df = pd.read_csv("CC GENERAL.csv")
x = df['CUST_ID']
y = df['PAYMENTS']
# 그래프 그리기
plt.figure(figsize=(8, 4)) # 캔버스(Figure) 크기 설정
            # 주가 변화 (가상)
plt.plot(x, y, label='label', color='orange', linestyle='-') # 선 그래프 그리기
plt.title('title') # 제목 / 시간에 따른 주가 추이
plt.xlabel('x_axis') # X축 레이블 / 시간 (일)
plt.ylabel('y_axis') # Y축 레이블 / 가격
plt.legend('bum') # 범례 표시
plt.grid(True) # 그리드 표시
plt.show() # 그래프 출력
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 데이터 준비 (사람들의 나이 분포 가정)
data_age = np.random.normal(loc=35, scale=10, size=500) # 평균 35, 표준편차 10
type(data_age)

# plt.figure(figsize=(8, 4))
# 히스토그램 그리기
plt.hist(data_age, bins=20, edgecolor='black', color='green', alpha=0.3) 
plt.title('title') # 데이터의 나이 분포 확인
plt.xlabel('x') # 나이
plt.ylabel('y') # 빈도수
plt.axvline(np.mean(data_age), color='red', linestyle='dashed', linewidth=1, label=f'평균: {np.mean(data_age):.2f}') # 평균선 추가
plt.legend()
plt.show()
```

```python
# 데이터 준비 (광고비 지출과 매출액 가정)
ad_spend = np.random.uniform(100, 1000, 50)
sales = ad_spend * 2.5 + np.random.normal(0, 500, 50) # 양의 상관관계

plt.figure(figsize=(8, 4))
# 산점도 그리기
plt.scatter(ad_spend, sales, color='purple', alpha=0.6, s=50) # s는 점의 크기
plt.title('title') # 광고 지출액과 매출액의 관계
plt.xlabel('x') # 광고 지출액 (독립변수)
plt.ylabel('y') # 매출액 (종속변수)
plt.show()
```
