```python
#----------------------------------
# 데이터 전처리
#----------------------------------
```

```python
# 라벨(정답,y값) 분리
y_train = train.pop("Credit_Score")
```

```python
y_train
```

```python
#----------------------------------
# 기존 레이블 인코딩
#----------------------------------
from sklearn.preprocessing import LabelEncoder

cols = train.columns[train.dtypes == object]

for col in cols:
    le = LabelEncoder()
    le.fit(pd.concat([train[col],test[col]] ,axis=0)) # 행합치기
    train[col] = le.transform(train[col])
    test[col] = le.transform(test[col])

# 범주형만 꺼내서 반복문으로 합침
# train과 test로 다시 나누기
```

```python
train
```

```python
test
```

```python
#----------------------------------
# 검증데이터 분할
#----------------------------------
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    train,
    y_train,
    test_size=0.2,
    random_state=0
)
```

```python
X_train.shape, X_val.shape, y_train.shape, y_val.shape
```

```python
#----------------------------------
# 머신러닝 학습 및 평가
#----------------------------------
```

```python
# 랜덤포레스트 학습
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(random_state=0) # 난수값 고정in-out 고정값
rf.fit(X_train, y_train) # 모델 학습
pred_rf=rf.predict_proba(X_val)
pred_rf
```

```python
# Lightgbm 학습
import lightgbm as lgb
lgbmc = lgb.LGBMClassifier(random_state=0, verbose=-1)
lgbmc.fit(X_train, y_train)
pred=lgbmc.predict(X_val) # 예측 y값
pre_proba = lgbmc.predict_proba(X_val)
```

```python
# 평가
```

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn.metrics import roc_auc_score

# roc_auc = roc_auc_score(y_val, pred[:,1])#pred[오답예측확률,오답예측확률]값
accuracy = accuracy_score(y_val,pred) # pred 예측 y값
f1 = f1_score (y_val, pred, average='macro') # pred 예측 y값
                           #(클래스별(1,2,3정답값) 성능을 동등하게 반영)

# print('roc_auc : ' , roc_auc)
print('accuracy : ' , accuracy)
print('f1 : ' , f1)


```
