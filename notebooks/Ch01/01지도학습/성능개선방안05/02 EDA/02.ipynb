카이제곱 검정 [범주형 X] - [범주형 Y]

```python
from scipy.stats import chi2_contingency
import pandas as pd

# 데이터 예시 (성별에 따라 제품 선택에 차이가 있도록 설정)
df = pd.DataFrame({
    'gender': ['M','M','M','M','M','F','F','F','F','F'],
    'product': ['A','A','A','B','B','B','B','B','A','A']
})
```

```python
# 교차표 생성
crosstab = pd.crosstab(df['gender'], df['product'])
crosstab
```

```python
# # 카이제곱 검정
chi2, pval, dof, expected = chi2_contingency(crosstab)
print('Chi-square: ',chi2)
print('p-value: ',pval)
print('자유도: ', dof)
print('기대빈도: ',expected)
```

```python
# # 결과 해석
alpha = 0.05
if pval < alpha:
    print('→ 두 변수는 유의한 관계가 있습니다.')
else:
    print('→ 두 변수는 독립적입니다.')
```

```python
import pandas as pd
# 1. CSV 파일 불러오기
print("CSV 파일 불러오기\n")
df = pd.read_csv("train.csv")
df.info()

# 2 target 분리
target = df.pop("income")
print("\n-----------------------------------------------------\n")
print("target 분리\n")
df.info()
print("\n-----------------------------------------------------\n")
print("범주형 x 분리\n")
# 3 범주형 x 분리
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()
categorical_cols
```

```python
# 4. 카이제곱 검정 수행 및 결과 저장
results = []

for col in categorical_cols:
    crosstab = pd.crosstab(df[col], target)
    chi2, p, dof, expected = chi2_contingency(crosstab)
    results.append({
        '변수명': col,
        'Chi2 통계량': round(chi2, 4),
        'p-value':round(p, 4),
        '자유도': dof,
        '유의성(p<0.05)': '유의함' if p < 0.05 else '유의하지 않음'
    })
print("-----------------------------------------------------\n")
print("DataFrame 생성 및 정렬\n")
# 6. DataFrame 생성 및 정렬
chi2_df = pd.DataFrame(results).sort_values('p-value', ascending=True).reset_index(drop=True)

print("-----------------------------------------------------\n")
print("결과 출력\n")
# 7. 결과 출력
print("=== 카이제곱 검정 결과 (income 종속변수 기준) ===")
display(chi2_df)
```
