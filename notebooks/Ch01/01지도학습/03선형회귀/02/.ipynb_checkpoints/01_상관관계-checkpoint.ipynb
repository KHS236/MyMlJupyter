```python
#--------------------------------------------
# 상관관계 확인 (수치형 )
#--------------------------------------------
import pandas as pd

    
data = {
    "키":[150,160,170,175,180],
    "몸무게":[40,50,60,70,80]
}

df = pd.DataFrame(data)
df

```

```python
# 상관계수 확인 (기본 : 피어슨 상관계수 -1 ~ 1 )
# +1에 가까울 수록 양의 상관관계
# -1에 가까울 수록 음의 상관관계
# 0에 가까울수록 상관이 거의 없음

correlation = df.corr()
print(correlation)
```

```python
# 피어슨 상관계수 (기본) 
# - 선형관계가 얼마나 강한지 확인 / 곡선관계 확인은 힘들다
print(df.corr())

# 스피어맨 상관계수
# - 순위(등수)를 기준으로 비교하는 방법 / 곡선관계도 확인 가능
print(df.corr(method='spearman'))

# 켄달타우 상관계수
# - 순위(등수)를 기준으로 비교하는 방법 (순서쌍을 만들어서 비교)
# / 곡선관계도 확인 가능
print(df.corr(method='kendall'))
```

```python
from scipy import stats
#피어슨 상관계수와 p_value 계산
print(stats.pearsonr(df['몸무게'],df['키']))

# 스피어맨 상관계수와 p-value 계산
print(stats.spearmanr(df['몸무게'],df['키']))

# 켄달타우 상관계수와 p-value 계산
print(stats.kendalltau(df['몸무게'],df['키']))

# statistic : 상관계수
# pValue : 유의확률( 관련이 있을 확률), 낮을수록 좋음
```
